### Строю графики

```python
df_img = df[['Itraxx Main Generic',
             'ER Itraxx Main Long',
             'ER Itraxx Main Short',
             'SPX']]

df_img = df_img.dropna()
df_img = df_img.set_index(( 'SPX',       'Date'))

df_img = df_img.drop(columns=[
    ( 'Itraxx Main Generic',       'Date'),
    ( 'ER Itraxx Main Long',       'Date'),
    ('ER Itraxx Main Short',       'Date')
])
df_img.index.names = ['Date']

df_img.columns = ['price', 'ER_Long', 'ER_Short', 'SnP_500_price', 'SnP_500_volume']

df_img # Теперь все выглядит куда красивее (и мы можем забыть о дате)
```

```python
plt.plot(df_img['price'])
plt.plot(df_img['ER_Long'])
plt.plot(df_img['ER_Short'])
```

### Исследую корреляции

#### Попарные

```python
display(df_img['price'].corr(df_img['ER_Long']))
display(df_img['price'].corr(df_img['ER_Short']))
display(df_img['ER_Long'].corr(df_img['ER_Short']))
```

Отсюда мы видим, что Long и Short по сути линейно "противоположны". Это видно и из графиков. На общем фоне выглядит как константа, но при более детальном рассмотрении можно получить какую-то информацию

```python
plt.plot(df_img['price'])
plt.plot((df_img['ER_Long'] + df_img['ER_Short']) / 2)
```

```python
plt.plot((df_img['ER_Long'] + df_img['ER_Short']) / 2)
```

#### Авто

```python
price_lags = []
ER_Long_lags = []
ER_Short_lags = []

for i in range(1000):
    price_lags.append(df_img['price'].autocorr(lag=i))
    ER_Long_lags.append(df_img['ER_Long'].autocorr(lag=i))
    ER_Short_lags.append(df_img['ER_Short'].autocorr(lag=i))
```

```python
plt.plot(price_lags) # blue
plt.plot(ER_Long_lags) # orange
plt.plot(ER_Short_lags) # green
```

Ну что тут сказать, цена весьма хаотичны, а индексе - менее.
