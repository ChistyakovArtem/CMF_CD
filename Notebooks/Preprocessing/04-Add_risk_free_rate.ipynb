```python
import pandas as pd
# source - "https://www.investing.com/rates-bonds/"+ 'country' + "-1-year-bond-yield-historical-data"
```

```python
pd.read_csv('03_result.csv')
```

```python
import os

big_df = pd.DataFrame()

for i in os.listdir('04-bond_yields'):
    df = pd.read_csv('04-bond_yields/' + i)
    country = i.split()[:2]
    if country[1][0] == '1':
        country = i.split()[0]
    else:
        country = "-".join(country)

    df = df.set_index('Date')
    df.index.names = ['']
    big_df[country + '_1_year_bonds_yield'] = df['Price']
```

```python
big_df = big_df[::-1]
```

```python
import calendar

ind = 0
month_map = {}
for i in calendar.month_abbr:
    month_map[i] = ind
    ind += 1
```

```python
tmp = big_df.index[100]

new_index = []
for tmp in big_df.index:
    month = tmp[:3]
    day = tmp[4:6]
    year = tmp[8:12]
    new_index.append(year + '-' + str(month_map[month]).rjust(2, '0') + '-' + day)
```

```python
big_df.index = new_index
big_df
```

```python
df = pd.read_csv('03_result.csv').set_index('Unnamed: 0')
df.index.names = ['']
df
```

```python
df = df.merge(big_df, left_index=True, right_index=True)
```
